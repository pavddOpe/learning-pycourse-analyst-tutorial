## Automatic File Sorter in File Explorer

```python
import os, shutil  # 'os' - operation system; 'shutil' - shell utilities - high-level operations on files.
```

```python
path = r"E:/Data Analytics/Beginner courses Data Bootcamp/Python/Temp/"    #'r' infront reads the path as str
```

```python
all_files = os.listdir(path)
all_files  # shows all the files under the previously described path
```

```python
path = r"E:/Data Analytics/Beginner courses Data Bootcamp/Python/Temp/"    #'r' infront reads the path as str

all_files = os.listdir(path)
all_files

folder_names = ['Excel project files', 'Wallpapers', 'Text files']

for loop in range(0,3):
    if not os.path.exists(path +folder_names[loop]):
        #print('Yes')
        os.makedirs(path + folder_names[loop])

for file in all_files:
    if '.xlsx' in file and not os.path.exists(path + 'Excel project files/' + file):
        shutil.move(path + file, path + 'Excel project files/' + file)
    elif '.jpg' in file and not os.path.exists(path + 'Wallpapers/' + file):
        shutil.move(path + file, path + 'Wallpapers/' + file)
    elif '.txt' in file and not os.path.exists(path + 'Text files/' + file):
        shutil.move(path + file, path + 'Text files/' + file)
```
