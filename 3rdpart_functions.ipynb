## Functions
*it's a block of code that runs only when being called*

```python
def just_func():
    print("It's just done :)")

just_func()
```

```python
def num_squared(number):   # "number" here it's an argument of the function; so that when we call it we can add argument inside the parentecies, so that the function (body) will work
    print(number**2)
num_squared(8)
```

```python
def num_squared_cust(number,power):   # we can use as many arguments as we want to
    print(number**power)
num_squared_cust(9,4)
```

#### *Arbiture arguments*

```python
args_tuple = (3,9,10,11,99,115,200,4,2,66)

def number_args(*number):
    print(number[0]*number[2])
number_args(*args_tuple)
```

```python
def num_squared_cust(number,power):   # this is 'key word argument' - we can have more control on arguments using this method
    print(number**power)
num_squared_cust(power = 9,number = 4)
```

```python
def number_kwarg(**number):  # "arbitory key word arguments" means that we don't know how many numbers gonna be there
    print('My number is: ' + number['integer'] + '. But your number is: ' + number ['integer2']+".")
number_kwarg(integer = '3041', integer2 = '452')
```
