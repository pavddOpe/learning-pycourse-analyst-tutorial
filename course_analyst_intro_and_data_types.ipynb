```python
#set - doesn't have duplicates; is unordered, so doesn't have index

daily_pints = {1,2,3}
type(daily_pints)
```

```python
daily_pints_log = {1,4,4,5,6,5,8,31,30,12,19} #having unique values!
print(daily_pints_log)
```

```python
weekly_pints_log = {1,8,33,45,56,31,30,12,19,30,48,51}
print(weekly_pints_log)
```

```python
print(daily_pints_log | weekly_pints_log) # '|' (straight line) shows the unique values when comparing two sets
```

```python
print(daily_pints_log & weekly_pints_log) #shows values that matches in comparing sets
```

```python
print(daily_pints_log - weekly_pints_log) #shows values that don't match
```

```python
print(daily_pints_log ^ weekly_pints_log) #shows completely unique values - that are either in one of comparing sets
```
