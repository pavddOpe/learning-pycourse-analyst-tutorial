#### *If-Elif-Else statement*

```python
if 25 > 10:
    print('It worked!')
```

```python
if 25 < 10:
    print('It worked!')
else:
    print("It didn't work...")
```

```python
if 25 < 10:
    print('It worked!') #we can do as many 'elif' statements as want
elif 25 <30:
    print('elif worked!')
else:
    print("It didn't work...")
```

```python
if (25 < 10) or (1<3): #we can also use logical operators here
    print('it worked!')
```

```python
print('It is your business') if 10 == 11 else print('It is not my business') 
```

```python
if (30 < 15) or (8 > 7):
    print('It works somehow...')
    if 10 > 5:
        print('Nested IF line works!') # we can create many nested IF statements in here
elif 35 < 20:
    print('You are gay')
elif 10 != 10:
    print('Definitely gay')
else:
    print('You are being gay for 10 years already!')
```

#### *For Loops*


```python
integers = [1,2,3,4,5] # below 'for number' corresponds to this list (as there are number in it)
# instead of list it cam be string, tuple etc.
```

```python
for number in integers:
    print(number) # it doesn't have to be exactly this variable again 'number'. It can be anything
```

```python
for boom in integers:
    print('oh yeah')
```

```python
for kek in integers:
    print(kek + kek)
```

```python
games_dict = {'Name': 'Metro', 'Part': 2, 'Favourite characters': ['Mykola', 'Pavlo', 'Ivan']}
```

```python
for game in games_dict.values():
    print(game)
```

```python
for key,value in games_dict.items():
    print(key, '->' , value)
```

#### Nested *For Loops*

```python
eyes_colors = ['blue','green','brown','dark']
eyes_size = ['big','small','middle','tiny']
```

```python
for one in eyes_size: # we can create different and many nested loops; they'll be iterating every time
    for two in eyes_colors:
        print('Their', one, two, 'eyes are amazing!')
```

#### *While loops*


```python
number = 0

while number <= 5: # it runs until conditions can't be met
    print(number)
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 3:
        break               # break statement stops the loop
    number = number + 1
```

```python
number = 0

while number < 5:
    print(number)
    if number == 6:
        break             
    number = number + 1
else:
    print('No longer <5')
```

```python
number = 0

while number < 5:
    print(number
    if number == 3:
        continue          #it will 'continue' the if (while) statement as long as code and memory allows, so will never stop.
    number = number + 1
else:
    print('No longer <5')
```
