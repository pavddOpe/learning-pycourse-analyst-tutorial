## HTML and Inspecting a Web Page

#### BeautifulSoup and Requests

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://hylogic.com/'
```

```python
portfolio = requests.get(url)  #checks if the path exists, if the source can be opened, if it is available
```

```python
soup = BeautifulSoup(portfolio.text, 'html')    #'.text' will bring us html code in text here...
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.find_all(class_ = 'menu-item')
```

#### Find and Find_all

```python
soup.find('div')  # it finds always the very first matching result
```

```python
soup.find_all('div', class_ = 'fade-in')
```

```python
soup.find_all(class_ = 'image fit')
```

```python
soup.find_all('h1')    # doesn't have attribute '.text' in it; it is only in '.find'
soup.find('h1').text
```

```python
soup.find('p').text.strip()
```
