## BMI Calculator 
*using SI - International System of Units measurement*

```python
# BMI = mass(kg)/height**2(m)
```

```python
name = input('Enter your name: ')

mass = int(input('Enter your weight in killograms: '))

height = float(input('Enter your height in meters: '))

BMI = (mass) / (height**2)

print(mass)
print(height)
print(BMI)

if BMI>0:                       # should be more that 0, if user put everything properly; else if not.
    if(BMI<18.5):
        print(name + ', you are underweight.')
    elif(BMI<=24.9):
        print(name + ', you are normal weight.')
    elif(BMI<=29.9):
        print(name + ', you are overweight.')
    elif(BMI<=34.9):
        print(name + ', you are obese.')
    elif(BMI<=39.9):
        print(name + ', you are severely obese.')
    else:
        print(name + ', you are morbidly obese.')
else:
    print('Please provide valid measurements.')     
```
